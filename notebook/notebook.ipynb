```python
from sklearn.preprocessing import LabelEncoder

cat_cols = [
    "time_of_day",
    "previous_day_mood",
    "face_emotion_hint",
    "ambience_type",
    "reflection_quality"
]

encoders = {}

for col in cat_cols:
    le = LabelEncoder()
    
    df[col] = df[col].astype(str)
    df_test[col] = df_test[col].astype(str)

    le.fit(list(df[col]) + list(df_test[col]))  # FIX

    df[col] = le.transform(df[col])
    df_test[col] = le.transform(df_test[col])

    encoders[col] = le
```

```python
meta_cols = [
    "sleep_hours",
    "stress_level",
    "energy_level",
    "duration_min",
    "time_of_day",
    "previous_day_mood",
    "face_emotion_hint",
    "ambience_type",
    "reflection_quality"
]

```

```python
from scipy.sparse import hstack, csr_matrix

X_meta_train = df[meta_cols].values.astype("float32")
X_meta_test = df_test[meta_cols].values.astype("float32")

X_meta_train_sparse = csr_matrix(X_meta_train)
X_meta_test_sparse = csr_matrix(X_meta_test)

X_train = hstack([x_text_train, X_meta_train_sparse])
X_test = hstack([x_text_test, X_meta_test_sparse])
```

model training


```python
from sklearn.preprocessing import LabelEncoder


le_state = LabelEncoder()
y_state = le_state.fit_transform(df["emotional_state"])

le_intensity = LabelEncoder()
y_intensity = le_intensity.fit_transform(df["intensity"])
```

```python
from sklearn.model_selection import train_test_split

X_tr, X_val, y_tr, y_val = train_test_split(
    X_train,
    y_state,
    test_size=0.2,
    random_state=42,
    stratify=y_state
)
```

```python
from xgboost import XGBClassifier

state_model = XGBClassifier(
    n_estimators=100,
    max_depth=6,
    learning_rate=0.1,
    random_state=42,
    eval_metric='mlogloss'
)

state_model.fit(X_tr, y_tr)
```

```python
y_pred = state_model.predict(X_val)
```

```python
from sklearn.metrics import accuracy_score

print("State Accuracy:", accuracy_score(y_val, y_pred))
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_val, y_pred))
```

```python
# ...existing code...

def decide_action(state, intensity, stress, energy, time_raw):

    if stress >= 8:
        return "breathing", "now"

    if time_raw == "night" and energy <= 4:
        return "sleep", "tonight"

    if state == "overwhelmed":
        if intensity >= 3:
            return "grounding", "now"
        else:
            return "breathing", "within_15_min"

    if state == "restless":
        if energy >= 6:
            return "movement", "now"
        elif intensity >= 3:
            return "grounding", "within_15_min"
        else:
            return "breathing", "within_15_min"

    if state == "mixed":
        return "journaling", "later_today"

    if state == "focused":
        if energy >= 6 and stress <= 5:
            return "deep_work", "now"
        else:
            return "light_planning", "within_15_min"

    if state == "calm":
        if time_raw == "night":
            return "sleep", "tonight"
        else:
            return "light_planning", "later_today"

    if state == "neutral":
        return "light_planning", "later_today"

    if energy <= 2:
        return "rest", "now"

    if energy <= 4:
        return "rest", "later_today"

    if energy >= 6 and time_raw == "night":   
        return "deep_work", "tomorrow_morning"

    return "light_planning", "later_today"

```

```python
def generate_message(state, action):
    
    if state == "overwhelmed":
        return f"You seem overwhelmed. Let’s slow things down—try {action}."

    if state == "restless":
        return f"You seem a bit restless. A small step like {action} can help you settle."

    if state == "focused":
        return f"You’re in a good state. This is a great time for {action}."

    if state == "calm":
        return f"You seem calm. You can continue with {action} at your pace."

    if state == "mixed":
        return f"You have mixed feelings. Try {action} to gain clarity."

    if state == "neutral":
        return f"You seem neutral. {action} could help you move forward."

    return f"Try {action} to improve your current state."
```

confidence and uncertainty

```python
#confidence
probs = state_model.predict_proba(X_test)
confidence = probs.max(axis=1)

#uncertainty
text_len = df_test["journal_text"].apply(lambda x: len(str(x).split()))
uncertain_flag = ((confidence < 0.6) | (text_len < 3)).astype(int)
```

```python

pred_state = state_model.predict(X_test)
pred_intensity = intensity_model.predict(X_test)


pred_state_labels = le_state.inverse_transform(pred_state)

stress_vals = df_test["stress_level"].values
energy_vals = df_test["energy_level"].values
time_vals = df_test["time_of_day_raw"].values


actions = []
times = []
messages = []

for i in range(len(pred_state_labels)):

    action, timing = decide_action(
        pred_state_labels[i],
        pred_intensity[i],
        stress_vals[i],
        energy_vals[i],
        time_vals[i]
    )

    msg = generate_message(pred_state_labels[i], action)

    actions.append(action)
    times.append(timing)
    messages.append(msg)



state_probs = state_model.predict_proba(X_test)
state_conf = state_probs.max(axis=1)

int_probs = intensity_model.predict_proba(X_test)
int_conf = int_probs.max(axis=1)

confidence = 0.7 * state_conf + 0.3 * int_conf

text_len = df_test["journal_text"].apply(lambda x: len(str(x).split()))
uncertain_flag = ((confidence < 0.6) | (text_len < 3)).astype(int)



output = pd.DataFrame({
    "id": df_test["id"],
    "predicted_state": pred_state_labels,
    "predicted_intensity": pred_intensity,
    "confidence": confidence,
    "uncertain_flag": uncertain_flag,
    "what_to_do": actions,
    "when_to_do": times,
    "message": messages
})



print(output.head(10))

output.to_csv("predictions.csv", index=False)
```
